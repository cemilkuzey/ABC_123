# Extracting ZIP Files and Loading Datasets

```python
# Bonus
# Import different file types from GitHub into Python

# Install library (run once if needed)
!pip install pyreadstat openpyxl

# STEP 1 — Import libraries
import urllib.request
import pandas as pd
import pyreadstat

# STEP 2 — Paste RAW GitHub link here
# RAW link format: https://raw.githubusercontent.com/<username>/<repository>/<branch>/<file_path>
url = "https://raw.githubusercontent.com/cemilkuzey/ABC_123/main/STATA_data.dta"

# STEP 3 — Get file name automatically
filename = url.split("/")[-1]

# STEP 4 — Download file
urllib.request.urlretrieve(url, filename)

# STEP 5 — Read file based on type
if filename.endswith(".csv"):
    df = pd.read_csv(filename)

elif filename.endswith(".xlsx"):
    df = pd.read_excel(filename)

elif filename.endswith(".json"):
    df = pd.read_json(filename)

elif filename.endswith(".sav"):
    df, meta = pyreadstat.read_sav(filename)

elif filename.endswith(".dta"):
    df, meta = pyreadstat.read_dta(filename)

elif filename.endswith(".sas7bdat"):
    df, meta = pyreadstat.read_sas7bdat(filename)

elif filename.endswith(".xpt"):
    df, meta = pyreadstat.read_xport(filename)

else:
    print("Unsupported file type")
    df = None

# STEP 6 — Show first rows (force output)
df.head()
```
